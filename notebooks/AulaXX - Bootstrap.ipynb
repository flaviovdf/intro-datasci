```python
from scipy import stats as ss

import matplotlib.pyplot as plt
import numpy as np
import pandas as pd

plt.ion()
```

```python
df = pd.read_csv('data/nba_salaries.csv')
```

```python
df
```

```python
N = 10000
idx = np.random.choice(df.index)
```

```python
S = 100
N = 10000
values = np.zeros(N)
for i in range(N):
    sample = np.random.choice(df.index, replace=True, size=S) # Escolhe 100 elementos
    values[i] = df['SALARY'][sample].mean()
```

```python
plt.hist(values, bins=20, normed=True);
plt.xlabel('Media da Amostra de Tamanho 100')
plt.ylabel('P(media)')
```

```python
values.mean()
```

```python
values.std()
```

```python
from IPython.display import HTML
from matplotlib import animation
```

```python
def update_hist(num, data):
    plt.cla()
    plt.hist(data[0:100 * (num+1)], bins=50, normed=True)
    plt.xlabel('Media da Amostra de Tamanho 100')
    plt.ylabel('Pr(media)')
```

```python
fig = plt.figure()
ani = animation.FuncAnimation(fig, update_hist, 30, fargs=(values, ))
HTML(ani.to_html5_video())
```
